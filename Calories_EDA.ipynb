```python
# Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read csv file of all activities
df=pd.read_csv('Activities (1).csv')
```

```python
#Checking first 5 rows
df.head()
```

```python
# Print detailed information about data
df.info()
```

```python
# Check basic statistical details
df.describe().transpose()
```

```python
#Number of null values
df.isnull().sum()
```

```python
#Selecting running activities among all types of activities
df=df[df['Activity Type']=='Running']
df
```

```python
#Converting date object to date-time object
df['Date']=df['Date'].apply(pd.to_datetime)
```

```python
df['Date']
```

```python
#Making separate columns for month, year and time
df['month']=df['Date'].apply(lambda x:x.month)
df['year']=df['Date'].apply(lambda x:x.year)
df['Activity Time'] = pd.DatetimeIndex(df['Date']).time
```

```python
#All max temp values are 0 hence drop the column
df[df['Max Temp']!=0]
df=df.drop(['Max Temp'],axis=1)
```

```python
# Print all columns to check parameters available
df.columns
```

```python
df.columns[16:26]
```

```python
#Drop columns as value is 0 (no significance)
df=df.drop(df.iloc[:,16:26],axis=1)
df=df.drop('Favorite',axis=1)
#Drop lap time as distance that counts as 1 lap setting changed multiple times
df=df.drop('Best Lap Time', axis=1)
```

```python
# Changed entries as "--" calories and treated them as 0
# df['Calories'].iloc[362]
df['Calories'].iloc[362]=0
```

```python
#Converting calories series from object to float and removing commas for values having thousandth digit and above
df['Calories']=pd.to_numeric(df['Calories'].apply(str).str.replace(',', ''))
```

```python
# Converting "--" entries to 0
df['Max Speed'].iloc[362]='0:0'
df['Avg Speed'].iloc[362]='0:0'
```

```python
#Converting max and avg speed from object to float 64 and changing from form min:sec/km to km/min i.e changing 6:30 min/km to 0.15km/min for convenience of analysis like higher the speed higher the calories burnt. In min/km lower the min/km value more will be the calories burnt

def speedmin(x):
    mm,ss=x.split(':')
    if(float(mm)==0 or float(ss)==0):
        return 0
    return 1/(float(mm)+float(ss)/60)
df['Max Speed']=df['Max Speed'].apply(speedmin)
df['Avg Speed']=df['Avg Speed'].apply(speedmin)
```

```python
#Converting Min Elevation to int type
df['Min Elevation']=pd.to_numeric(df['Min Elevation'])
df['Max Elevation']=pd.to_numeric(df['Max Elevation'])
```

```python
#Dropping rows with Avg Run Cadence and Max Run Cadence as --
df=df[df['Avg Run Cadence']!='--']
df=df[df['Max Run Cadence']!='--']
```

```python
#Converting Avg Run Cadence and Max Run Cadence to int
df['Avg Run Cadence']=pd.to_numeric(df['Avg Run Cadence'])
df['Max Run Cadence']=pd.to_numeric(df['Max Run Cadence'])
```

```python
#Function to convert run time to seconds
def tosec(x):
    hh, mm, ss = x.split(':')
    return float(hh) * 3600 + float(mm) * 60 + float(ss)
df['Time in sec']=df['Time'].apply(tosec)

#Applying the function to Moving Time and Elapsed Time
df['Moving Time in sec']=df['Moving Time'].apply(tosec)
df['Elapsed Time in sec']=df['Elapsed Time'].apply(tosec)
```

```python
df=df.drop(['Number of Laps'], axis=1)
```

```python
# 'Distance','Time in sec','Elapsed Time in sec showed max correlaton with calories hence choosing these parameters
X=df[['Distance','Time in sec','Elapsed Time in sec']]
#X=df[['Distance','Time in sec']]
y=df['Calories']
```

```python
#Import functions from scikit learn
from sklearn.model_selection import train_test_split
```

```python
#Splitting training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
#Fitting the model on training data
lm.fit(X_train,y_train)
```

```python
lm.intercept_
```

```python
lm.coef_
#These are regression coefficients completely different from correlation coeffs
```

```python
cdf=pd.DataFrame(lm.coef_,index=X.columns,columns=['Coeff'])
```

```python
# 1 km increase in distance results in 65.95 cal increase
#1 sec increase in time results in 0.007 cal increase       
cdf
```

```python
#Predictions
predictions=lm.predict(X_test)
```

```python
#True output of calories
y_test
```

```python
#Scatterplot of true values and predictions
sns.scatterplot(predictions,y_test)
```

```python
#Residuals/Error has a normal distribution
sns.distplot(y_test-predictions)
```

```python
#Importing functions for metrics
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test,predictions)
```

```python
metrics.mean_squared_error(y_test,predictions)
```

```python
#Prediction error
np.sqrt(metrics.mean_squared_error(y_test,predictions))
```
